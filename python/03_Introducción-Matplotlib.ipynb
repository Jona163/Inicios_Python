## Representación gráfica de datos.
Si la función de trazado se le da a una matriz de datos, la usará como coordenadas en el eje vertical, utilizará el índice de cada punto de datos en el array como la coordenada horizontal.

```python
plt.plot([1, 2, 3, 4, 5, 7, 8, 3, 1])
plt.show()
```

Tambien sepuede proporcionar ds matrices: una para el eje horizontal y otra para el eje vertical.

```python
plt.plot([-3, -1, 0, 4, 7], [1, 4, 6, 7, 8])
plt.show()
```

Pueden modificarse las longitudes de los ejes para que la figura no se vea tan ajustada

```python
plt.plot([-3, -1, 0, 4, 7], [1, 4, 6, 7, 8])
plt.axis([-4, 8, 0, 10]) # [ xmin, xmax, ymin, ymax]
plt.show()
```

Se sige el mismo procedimiento para pintar una función matemática

```python
import numpy as np
x = np.linspace(-2, 2, 500)
y = x**2

plt.plot(x, y)
plt.show
```

Tambien puede modificarse el estilo de la gráfica para que contenga más información.

```python
plt.plot(x, y)
plt.title("Square Funtion")
plt.xlabel("chido")
plt.ylabel("Y = x**2")
plt.grid(True)
plt.show()
```

Pueden superponerse gráficas y cambiar el estilo de las lineas

```python
import numpy as np
x = np.linspace(-2, 2, 500)
y = x**2
y2 = x + 1

plt.plot(x, y, 'b--', x, y2, 'g')
plt.show()
```

Para poder diferneciar entre ambas funciones, siempre es recomendado añadir una leyenda.

```python
import numpy as np
x = np.linspace(-2, 2, 500)
y = x**2
y2 = x + 1

plt.plot(x, y, 'b--', label = "x**2")
plt.plot(x, y2, 'g', label = "x + 1")
plt.legend(loc = "best")
plt.grid(True)
plt.show()
```

Para que las gráficas no queden tan ajustadas, se puede hacer la figura más grande.

```python
plt.figure(figsize = (14, 6))

plt.subplot(1, 2, 1) # 1 Rows, 2 Columns, 1st Subplot.
plt.plot(x, y, 'b--')
plt.xlabel("x1", fontsize = 14)
plt.xlabel("y2", fontsize = 14)
plt.grid(True)

plt.subplot(1, 2, 2) # 1 Rows, 2 Columns, 2st Subplot.
plt.plot(x, y2, 'g')
plt.xlabel("x1", fontsize = 14)
plt.xlabel("y2", fontsize = 14)
plt.grid(True)
```

### Scatter Plots

```python
from numpy.random import rand
x, y = rand(2, 100)
plt.scatter(x, y)
plt.show()
```

```python
from numpy.random import rand
x, y = rand(2, 100)
x2, y2 = rand(2, 100)
plt.scatter(x, y, c = 'red')
plt.scatter(x2, y2, c = 'blue')
plt.show()
```

## Histogramas

```python
data = [1, 1.1, 1.8, 2, 2.1, 3.2, 3, 3, 3, 3]
plt.subplot(211) # 2 Rows, 1 Column, ist Subplot
plt.hist(data, bins = 10, rwidth = 0.8)

plt.subplot(212) # 2 Rows, 1 Column, 2st Subplot
plt.hist(data, bins = [1, 1.5, 2, 2.5, 3], rwidth = 0.95)
plt.xlabel("Valor")
plt.ylabel("Frecuencia")

plt.show()
```

## Guardar la Figuras

```python
import numpy as np
from numpy.random import rand
x, y = rand(2, 100)
x2, y2 = rand(2, 100)

plt.scatter(x, y, c = 'red')
plt.scatter(x2, y2, c = 'blue')
plt.legend(loc = "best")
plt.savefig("Mi_Gráfiquita.png", transparent = True)
```
